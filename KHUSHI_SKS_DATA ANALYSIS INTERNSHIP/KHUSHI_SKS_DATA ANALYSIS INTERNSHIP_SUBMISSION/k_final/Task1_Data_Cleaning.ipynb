# Task 1 - Data Cleaning and Preprocessing
**Dataset:** Telco Customer Churn  

In this task, I performed data cleaning and preprocessing on the Telco Customer Churn dataset.

The objective was to prepare the raw dataset for analysis and model building.

```python
# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print("libraries loaded successfully")
```

## Step 1: Load and Inspect the Data
From the initial inspection, I observed that some columns had incorrect data types.  

```python
df = pd.read_csv('Telco_Customer_Churn_Dataset_.csv')
print("Dataset shape:", df.shape)
print()
df.head()
```

```python
df.info()
```

```python
# checking basic stats for numeric columns
df.describe()
```

## Step 2: Fix the TotalCharges Column
I noticed that the 'TotalCharges' column was stored as object type.So I converted it to numeric for proper analysis.  

```python
# TotalCharges should be numeric but pandas read it as text
# check what values are causing this
print("Unique problematic values in TotalCharges:")
print(df[df['TotalCharges'] == ' '][['customerID', 'tenure', 'MonthlyCharges', 'TotalCharges']].head(10))
```

```python
#  blank spaces ' ' in TotalCharges for new customers (tenure = 0)
# these customers just signed up so they haven't been charged yet
# replacing blanks with NaN first, then converting to float

df['TotalCharges'] = df['TotalCharges'].replace(' ', np.nan)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

print("TotalCharges dtype now:", df['TotalCharges'].dtype)
print("Missing values in TotalCharges:", df['TotalCharges'].isnull().sum())
```

## Step 3: Handle Missing Values
There were missing values in the dataset.  
I handled them by:  
- Removing rows where necessary  
- Filling values using appropriate methods.  

```python
# checking all columns for missing values
print("Missing values per column:")
print(df.isnull().sum())
print()
print("Total missing:", df.isnull().sum().sum())
```

```python
# filling TotalCharges NaN with 0 because these are new customers with no charges yet
# NOT using mean because that would be incorrect for this specific case
df['TotalCharges'].fillna(0, inplace=True)

print("Missing values after fix:", df.isnull().sum().sum())
print()

# also checking for duplicate rows
print("Duplicate rows:", df.duplicated().sum())
print("Unique customers:", df['customerID'].nunique())
```

## Step 4: Encode the Target Column (Churn)
The 'Churn' column contains 'Yes' and 'No' values.   
Since machine learning models require numerical inputs, I converted them into 1 (Yes) and 0 (No).  

After converting, I was also able to calculate the churn rate easily using the mean of the column  

```python
# converting Churn from Yes/No to 1/0
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})

print("Churn value counts after encoding:")
print(df['Churn'].value_counts())
print()
print("Overall churn rate:", round(df['Churn'].mean() * 100, 2), "%")
```

## Step 5: One-Hot Encoding for Categorical Variables
The dataset contains categorical values like 'Month-to-month', 'Female', etc. 
Since machine learning models require numerical input, I converted these columns into numerical format.

I used One-Hot Encoding because these categories do not have any order. 
Using label encoding could create an artificial ranking between categories, which may affect the model.

```python
# checking for categorical columns
cat_cols = df.select_dtypes(include='object').columns.tolist()
cat_cols.remove('customerID')  # customerID is just an ID, not a feature
print("Categorical columns to encode:", cat_cols)
print()
print("Unique values in each:")
for col in cat_cols:
    print(f"  {col}: {df[col].unique()}")
```

```python
# making a copy for ML  keeping original df clean for visualisations
# the original has readable labels like 'Month-to-month' which are better for charts
df_encoded = df.copy()
df_encoded.drop(columns=['customerID'], inplace=True)

# one-hot encoding
df_encoded = pd.get_dummies(df_encoded, drop_first=True)

# newer pandas versions sometimes produce bool columns instead of int fixing that
bool_cols = df_encoded.select_dtypes(include='bool').columns
df_encoded[bool_cols] = df_encoded[bool_cols].astype(int)

print("Shape before encoding:", df.shape)
print("Shape after encoding:", df_encoded.shape)
print()
print("All columns are numeric now:", df_encoded.select_dtypes(include='object').shape[1] == 0)
```

## Step 6: Final Check and Save
Finally, I verified that the dataset has no missing values and correct data types before saving it.

```python
print("=== Final Data Quality Check ===")
print()
print("Shape:", df_encoded.shape)
print("Missing values:", df_encoded.isnull().sum().sum())
print("Duplicate rows:", df_encoded.duplicated().sum())
print()
print("Sample of first few columns:")
print(df_encoded.iloc[:3, :8])
```

```python
# saving two files:
# cleaned_churn_data.csv  human readable with original text labels (for EDA/visualisation)
# ml_ready_churn_data.csv fully encoded numeric version (for ML model)

df.to_csv('cleaned_churn_data.csv', index=False)
df_encoded.to_csv('ml_ready_churn_data.csv', index=False)

print("Saved: cleaned_churn_data.csv ->", df.shape)
print("Saved: ml_ready_churn_data.csv ->", df_encoded.shape)
print()
print("Task 1 complete!")
```
